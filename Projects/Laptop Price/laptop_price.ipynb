```python
ldf.rename(columns={'sex':'gender'},inplace=True)
ldf.rename(columns={'processor_name':'processor'},inplace=True)
ldf.rename(columns={'processor_gnrtn':'gen'},inplace=True)
ldf.rename(columns={'graphic_card_gb':'gpu'},inplace=True)
ldf.rename(columns={'ram_gb':'ram'},inplace=True)
```

```python
ldf.drop(columns=['weight','ram_type'],inplace=True)
```

```python
num=ldf.select_dtypes(exclude='object')
```

```python
num
```

```python
char=ldf.select_dtypes(include='object')
```

```python
char
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
char['brand'] = le.fit_transform(char['brand'])
char['processor_brand'] = le.fit_transform(char['processor_brand'])
char['processor'] = le.fit_transform(char['processor'])
char['gen'] = le.fit_transform(char['gen'])
char['ram'] = le.fit_transform(char['ram'])
char['ssd'] = le.fit_transform(char['ssd'])
char['hdd'] = le.fit_transform(char['hdd'])
char['os'] = le.fit_transform(char['os'])
char['os_bit'] = le.fit_transform(char['os_bit'])
char['gpu'] = le.fit_transform(char['gpu'])
char['warranty'] = le.fit_transform(char['warranty'])
char['Touchscreen'] = le.fit_transform(char['Touchscreen'])
char['msoffice'] = le.fit_transform(char['msoffice'])
char['rating'] = le.fit_transform(char['rating'])
```

```python
char
```

```python
clean_data=pd.concat([char,num],axis=1)
```

```python
clean_data
```

```python
clean_data.to_csv('new_Cleaned_Laptop_Price.csv', index=False)
```
