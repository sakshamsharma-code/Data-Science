 * it is a popular Python library for creating static, animated, and interactive visualizations, such as line plots, bar charts, scatter plots, histograms, and more.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

x=[2,3,4,5,2]
y=[1,2,5,6,1]  #used to define the x and y data points for plotting
plt.plot(x, y)  #used to plot the x and y data points
plt.show()  #used to display the plot
```

```python
xpoint=np.array([0,6])
ypoint=np.array([0,250])
plt.plot(xpoint, ypoint,'o')  #used to plot the x and y data points with 'o' marker
plt.show()  #used to display the plot
```

```python
xpoint=np.array([1,8,5,9])
plt.plot(xpoint,marker= 'o')  #used to plot the x data points with 'o' marker or scatter point
plt.show()  #used to display the plot
```

```python
xpoint=np.array([1,8,5,9])
plt.plot(xpoint,marker= '*')  
plt.show()
```

```python
xpoint=np.array([1,8,5,9])
plt.plot(xpoint,'o:b')  
plt.show()
```

```python
xpoint=np.array([1,8,5,9])
plt.plot(xpoint,marker= 'o', ms=38)  #used to plot the x data points with 'o' marker and set marker size to 38
plt.show()
```

```python
xpoint=np.array([1,8,5,9])
plt.plot(xpoint,marker= '*', ms=38)  
plt.show()
```

```python
xpoint=np.array([1,8,5,9])
plt.plot(xpoint,marker= '*', ms=38, mec='red')  
plt.show()
```

```python
xpoint=np.array([1,8,5,9])
plt.plot(xpoint,marker= '*', ms=38, mec='blue',mfc='pink')   #mec= marker edge color, mfc= marker face color
plt.show()
```

```python
x=np.array([1,8,5,7])
plt.plot(x,ls='dotted')
plt.show()  #used to display the plot with dotted line style
```

```python
x=np.array([1,8,5,7])
plt.plot(x,ls='dashed')  #used to display the plot with dashed line style
plt.show() 
```

```python
x=np.array([1,8,5,7])
plt.plot(x,linewidth=5)  #used to display the plot with line width set to 5
plt.show() 
```

```python
x=np.array([1,5,8,4])
y=np.array([2,3,9,5])
plt.plot(x)
plt.plot(y)
plt.show() 
```

```python
x=np.array([1,8,5,7])
plt.title('demo graph')  #used to set the title of the plot
plt.xlabel('x-axis')  #used to set the label for x-axis
plt.ylabel('y-axis')  #used to set the label for y-axis
plt.plot(x,linewidth=5) 
plt.show() 
```

```python
x=np.array([1,8,5,7])
plt.title('demo graph')  
plt.xlabel('x-axis')  
plt.ylabel('y-axis')  
plt.plot(x,linewidth=5) 
plt.grid()  #used to add a grid to the plot
plt.show() 
```

```python
x=np.array([1,5,8,4])
y=np.array([2,3,9,5])
plt.plot(x)
plt.plot(y)
plt.grid()
plt.show() 
```

```python
x=np.array([1,2,3,4])
y=np.array([1,4,8,6])
plt.subplot(1,2,1)  #used to create a subplot in a 1x2 grid, first subplot
plt.plot(x,y)
x=np.array([1,2,3,4])
y=np.array([1,5,9,4])
plt.subplot(1,2,2)  #used to create a subplot in a 1x2 grid, second subplot
plt.plot(x,y)
plt.show()  #used to display the subplots
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
c = np.array(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'red', 'green', 'blue', 'yellow', 'cyan'])
plt.scatter(x, y, s = sizes, c =c,alpha=0.5)  #used to create a scatter plot with sizes and colors
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
c = np.array(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'red', 'green', 'blue', 'yellow', 'cyan'])
plt.scatter(x, y, s = sizes, c =c,alpha=0.5,cmap='coolwarm')  
plt.colorbar()  #used to add a color bar to the scatter plot
plt.show()
```

# pie plot

```python
x=np.array([1,2,3,4])
plt.pie(x)  #used to create a pie chart of x data points
plt.show()  #used to display the pie chart
```

```python
x=np.array([1,2,3,4])
labels=['apple', 'banana', 'cherry', 'date']
plt.pie(x,labels=labels)  #used to create a pie chart of x data points with labels
plt.show()
```

```python
x=np.array([1,2,3,4])
labels=['apple', 'banana', 'cherry', 'date']
myexplode=[0,0.2,0,0]  #used to create an explode effect for the second slice
plt.pie(x,labels=labels,explode=myexplode) 
plt.show()
```

```python
x=np.array([1,2,3,4])
labels=['apple', 'banana', 'cherry', 'date']
myexplode=[0,0.2,0,0]
plt.pie(x,labels=labels,explode=myexplode,shadow=True)  #used to create a pie chart of x data points with labels and shadow effect
plt.show()
```

```python
x=np.array([1,2,3,4])
labels=['apple', 'banana', 'cherry', 'date']
myexplode=[0,0.2,0,0]
plt.pie(x,labels=labels,explode=myexplode,shadow=True,startangle=90)  
plt.show()
```

```python
x=np.array([1,2,3,4])
labels=['apple', 'banana', 'cherry', 'date']
myexplode=[0,0.2,0,0]
c= ['red', 'green', 'blue', 'yellow']
plt.pie(x,labels=labels,explode=myexplode,shadow=True,startangle=90,colors=c,autopct='%1.1f%%')  #used to create a pie chart of x data points with labels, shadow effect, start angle, colors, and percentage display
plt.show()
```

```python
x=np.array([1,2,3,4])
labels=['apple', 'banana', 'cherry', 'date']
myexplode=[0,0.2,0,0]
c= ['red', 'green', 'blue', 'yellow']
plt.pie(x,labels=labels,explode=myexplode,shadow=True,startangle=90,colors=c,autopct='%1.1f%%')  
plt.legend(title='my pie chart')
plt.show()
```
