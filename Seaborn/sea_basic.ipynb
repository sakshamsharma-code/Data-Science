## ðŸ” Relational Plots

### Basic Relplot
```python

plt.show()
```

### Colored by Time
```python
sns.relplot(x='total_bill', y='tip', data=data, hue='time')
plt.show()
```

### Styled by Smoker, Colored by Time
```python
sns.relplot(x='total_bill', y='tip', data=data, style='smoker', hue='time')
plt.show()
```

### Split by Time (Rows)
```python
sns.relplot(x='total_bill', y='tip', data=data, style='smoker', kind='scatter', row='time', hue='time')
plt.show()
```

### Split by Smoker (Columns)
```python

plt.show()
```

### Split by Day (Columns)
```python

plt.show()
```

## ðŸ–ï¸ Custom Titles and Labels

```python

plt.show()
```
