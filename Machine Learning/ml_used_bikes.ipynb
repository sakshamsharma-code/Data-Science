```python
import pandas as pd
```

```python
mdf=pd.read_csv('new_Cleaned_Used_Bikes.csv')
```

```python
mdf
```

```python
mdf.dropna(inplace=True)
x = mdf.drop('price', axis=1)
y = mdf[['price']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_train, y_train)
```

```python
lr.score(x_test, y_test) # Evaluate the model accuracy
```

```python
pred=lr.predict(x_test)
```

```python
pred
```

```python
y_test
```

```python
pred_df = pd.DataFrame(pred, columns=["Predicted_Price"]).reset_index(drop=True)
y_test_reset = y_test.reset_index(drop=True)
result = pd.concat([pred_df, y_test_reset], axis=1)
```

```python
result
```
